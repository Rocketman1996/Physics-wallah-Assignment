Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# let's load the tips library
tips = sns.load_dataset("tips")
tips
```

```python
# Histograms:

# Use: Histograms are used to visualize the distribution of a single numerical variable.
# They display the frequency or count of data points within predefined bins or intervals.

a = sns.histplot(data = tips,bins=7, kde=True, color='skyblue')
```

```python
# Scatter Plots:

# Use: Scatter plots are useful for visualizing the relationship between two numerical variables. They help identify patterns, correlations, and outliers in the data.
b = sns.scatterplot(data = tips, x = tips.total_bill, y = tips.tip, hue = 'smoker')
```

```python
# seaborn.relplot() is a function in the Seaborn library that is used for creating scatter plots and line plots for exploring the relationship between two numerical variables

t = sns.relplot(data = tips, x = tips.total_bill, y = tips.tip, hue = 'smoker', style = "smoker")
```

```python
# Bar Plots:

# Use: Bar plots are used to display and compare the values of categorical data.
# They are great for showing the distribution of a categorical variable or comparing differences.

c  = sns.barplot(tips, x = tips.sex, y = tips.tip, hue = "smoker")
```

```python
# Box Plots (Box-and-Whisker Plots):

# Use: Box plots are helpful for visualizing the distribution, spread, and outliers in numerical data. They provide a summary of the data's central tendency and spread.
d = sns.boxplot(data = tips, x = tips.sex, y = tips.total_bill)
# So total bill starts a bit lower for females
# the median is somewhere at 18 or 19
# Total bill is max for male and a bit less for female for about 5
# we have some few outliers for male which starts from 40 to 50 for mens and from 30 to about 45 for females
# on general average total_bill for men is 40 and 30 for female.
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python

dat1 = sns.load_dataset("titanic")
plt.figure(figsize=(12, 5))  # Adjust the figure size (optional)

# Create the first box plot for 'pclass' vs 'age'
plt.subplot(1, 2, 1)  # Create the first subplot
sns.boxplot(data=dat1, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')

# Create the second box plot for 'pclass' vs 'fare'
plt.subplot(1, 2, 2)  # Create the second subplot
sns.boxplot(data=dat1, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
dat4 = sns.load_dataset("flights")
dat4
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a pivot table to reshape the data for the heatmap
pivot_table = dat4.pivot_table(index='month', columns='year', values='passengers')


sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Flights Data Heatmap')
plt.show()
```
